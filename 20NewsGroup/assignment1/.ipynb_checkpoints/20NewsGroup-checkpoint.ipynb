```python
import pandas as pd
import shutil
import glob
```

# Step 2

## create classes

+ create computer class and add all files of sub category folders from 20news-bydate-test/train

```python
count = 0

for filename in glob.iglob('20news-bydate*/comp.*/*', recursive=True):
    shutil.copy(filename, 'classes/computer')
    count += 1
    print(filename)
    
print("There are ", count , " files add ")
```

+ create science class and add all files of sub category folders from 20news-bydate-test/train

```python
count = 0

for filename in glob.iglob('20news-bydate*/sci.*/*', recursive=True):
    shutil.copy(filename, 'classes/science')
    count += 1
    print(filename)
    
print("There are ", count , " files add ")
```

+ create recreational class and add all files of sub category folders from 20news-bydate-test/train

```python
count = 0

for filename in glob.iglob('20news-bydate*/rec.*/*', recursive=True):
    if filename == glob.iglob('classes/recreational/filename', recursive=False)
        count -=1
        
    shutil.copy(filename, 'classes/recreational')
    count += 1
    print(filename)
    
print("There are ", count , " files add ")
```
